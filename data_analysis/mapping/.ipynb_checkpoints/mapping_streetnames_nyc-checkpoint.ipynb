````python
def convert_geojson(df, filename):
    df = df.copy()
    df['geometry'] = df['Coordinates'].apply(wkt.loads)
    gdf = gpd.GeoDataFrame(df, geometry='geometry')
    gdf.crs = "EPSG:4326"
    gdf.to_file(filename, driver='GeoJSON')
````

````python
convert_geojson(df, "honorary_nyc_sts.geojson")
````

````python
import pandas as pd
import geopandas as gpd
from shapely import wkt

# Step 1: Load the CSV
df = pd.read_csv('mapping_nyc.csv')

# Step 2: Convert 'geometry' column from WKT string to shapely geometry
df['geometry'] = df['Coordinates'].apply(wkt.loads)

# Step 3: Convert to GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry='geometry', crs='EPSG:4326')

# Step 4: Calculate centroid (midpoint) of each geometry
gdf['centroid'] = gdf.geometry.centroid

# Step 5: Extract lat/lon from centroid
gdf['latitude'] = gdf.centroid.y
gdf['longitude'] = gdf.centroid.x

# Step 6: Save to CSV
gdf.drop(columns=['centroid']).to_csv('mapping_midpoints.csv', index=False)

print("✅ Done! Saved midpoint coordinates to 'nyc_streets_with_midpoints.csv'")
````

````python
gdf[gdf["Membership"] == "Male"].to_csv('male_midpoints.csv', index=False)
````

````python
gdf[gdf["Membership"] == "Female"].to_csv('female_midpoints.csv', index=False)
````

````python
# misc, group, male and female
gdf[(gdf["Membership"] != "Female") & (df["Membership"] != "Male")].to_csv('other_midpoints.csv', index=False)
````

## Adding additional data for tooltips 
using data we got from ```hyc_honorary_streenames.ipynb```

````python
female = pd.read_csv("facet_midpoints/female_midpoints.csv")
male = pd.read_csv("facet_midpoints/male_midpoints.csv")
````

````python
stats = pd.read_csv("../crosstab_borough.csv")
stats.drop(columns = "All", inplace = True)
````

````python
stats
````

````python
# creating dictionaries of the stats
female_named_pct = stats.iloc[0][1:].to_dict()
male_named_pct = stats.iloc[2][1:].to_dict()
````

````python
# getting population from censusreporter for each borough acc to ACS 2023

female_pop = {
    'Bronx': '53%',
    'Brooklyn': '52%',
    'Manhattan': '52%',
    'Queens': '51%',
    'Staten Island': '51%'}

male_pop = {
    'Bronx': '47%',
    'Brooklyn': '48%',
    'Manhattan': '48%',
    'Queens': '49%',
    'Staten Island': '49%'}
````

````python
female['pct_female_named_streets'] = female['Borough'].map(female_named_pct)
female["pct_female_pop"] = female["Borough"].map(female_pop)
female.head()
````

````python
male['pct_male_named_streets'] = male['Borough'].map(male_named_pct)
male["pct_male_pop"] = male["Borough"].map(male_pop)
male.head()
````
